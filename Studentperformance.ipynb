```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('StudentsPerformance.csv')
df.head()







```

```python
print(df.shape)
```

```python
print(df.info)
```

```python
print(df.describe)
```

```python
print(df.columns)
```

```python
df['total_score']=df[['math score','reading score','writing score']].sum(axis=1)
df['average_score']=df['total_score']/3
df.head()
```

```python
df['gender'].value_counts()
```

```python
df.groupby('gender')['average_score'].mean()
```

```python
plt.hist(df['total_score'],bins=10,color='blue',edgecolor='black')
plt.title("Total score")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python


plt.figure(figsize=(8,6))
plt.scatter(df['math score'],df['reading score'], color='skyblue', alpha=0.7)
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.grid()
plt.title('Math Score Distribution')

plt.show()
```

```python
df.boxplot(column='total_score',by='lunch')
plt.title("Total score")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python

gender_counts = df['gender'].value_counts()
labels = gender_counts.index
sizes = gender_counts.values

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.axis('equal')
plt.show()
```
